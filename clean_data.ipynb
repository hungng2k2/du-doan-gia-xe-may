```python
import pandas as pd
import os
```

```python
def clean_dataset(data):
    data['Năm đăng ký'] = data['Năm đăng ký'].str.replace('trước năm ','')
    data['Năm đăng ký'] = data['Năm đăng ký'].astype('Int64')

    data['Giá'] = data['Giá'].str.replace('đ','')
    data['Giá'] = data['Giá'].str.replace('.','')
    data['Giá'] = data['Giá'].astype('Int64')
    return data
```

Clean big dataset

```python
data = pd.read_csv('./raw_data/bigdataset.csv')
data.head()
```

```python
data = clean_dataset(data)
data.head()
```

```python
data.to_csv('./clean_data/bigdataset.csv', index=False)
```

Clean small dataset

```python
data = pd.read_csv('./raw_data/smalldataset.csv')
data.head()
```

```python
data = clean_dataset(data)
data.head()
```

```python
data.to_csv('./clean_data/smalldataset.csv', index=False)
```

```python
df = data.copy()
df.head()
```

```python
df_new = df.drop(['Tình trạng', 'Xuất xứ', 'Chính sách bảo hành'], axis=1)
df_new.head()
```

```python
df_new = pd.get_dummies(df_new, columns= ['Loại xe', 'Dung tích xe', 'Hãng xe', 'Dòng xe'])
```

```python
df_new.head()
```

```python
df_new = df_new.drop(['Giá'], axis=1)
df_new.head()
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=30)
```

```python
principal_components = pca.fit_transform(df_new)
```

```python
principal_df = pd.DataFrame(data=principal_components)
```
